```python
import requests
from bs4 import BeautifulSoup
```

```python
req_param = {"sid": 'game'}  # 'game', 'movie', 'tv', 'comics', 'tech'

url = "https://kr.ign.com/article/review/?keyword__type={sid}".format(**req_param)

# 요청 헤더 설정 : 브라우저 정보 (사람처럼 보이게 하려고 설정하는 것임)
req_header = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"
}
response = requests.get(url, headers=req_header)

if response.ok:
    response.encoding = 'utf-8' 
    # 응답 (response)에서 text 추출 - source 보기
    html = response.text
    # BeautifulSoup 객체 생성
    soup = BeautifulSoup(html, "html.parser")
    # caption = soup.select("span.caption")
    a_tag_list = soup.select("div.m a[href*='review']")

    # '#disqus_thread'을...
    # 1. list comprehension으로 제외
    filtered_a_tag_list = [tag for tag in a_tag_list if '#disqus_thread' not in tag['href']]
    # 2. 또는 정규 표현식(regex)로 제외

    for a_tag in filtered_a_tag_list:
        span_elem = a_tag.find("span", class_="caption")
        if span_elem:
            title = span_elem.get_text(strip=True)
        else:
            title = "(No caption)"
        link = a_tag["href"]
        print(f"제목: {title}\n링크: {link}\n")
else:
    # 응답(response)이 Error 이면 status code 출력
    print(f"Error Code = {response.status_code}")
```
